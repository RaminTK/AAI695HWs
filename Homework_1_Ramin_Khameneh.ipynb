```python
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_squared_error
import numpy as np 
```

### Part 1) Please plot the noisy data and the polynomial you found (in the same figure). You can use any value of m selected from 2, 3, 4, 5, 6.  

Plot for the noisy data

```python
plt.style.use('seaborn-whitegrid') 
noise_scale = 100 
number_of_samples = 50 
x =  30*(np.random.rand(number_of_samples, 1) - 0.5) 
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
plt.plot(x,y,'ro') 
```

polynomial for m = 2

```python
x = x.reshape(1,len(x))[0,:]
y = y.reshape(1,len(y))[0,:]
coeff = np.polyfit(x,y,2)
xn = np.linspace(-15,15,100)
yn = np.poly1d(coeff)
plt.plot( xn,yn(xn),x,y,'ro')
```

polynomial for m = 3

```python
coeff = np.polyfit(x,y,3)
xn = np.linspace(-15,15,100)
yn = np.poly1d(coeff)
plt.plot( xn,yn(xn),x,y,'ro')
```

polynomial for m = 4

polynomial for m = 5

### Part 2) Plot MSE versus order m, for m = 1, 2, 3, 4, 5, 6, 7, 8 respectively. Identify the best choice of m. 

```python
for i in range(1,9):
  p = np.polyfit(x,y,i)
  y_p = np.polyval(p,x)
  print('m = ', i, 'MSE = ',  mean_squared_error(y,y_p))
```

The best m is **m = 8** beacuse it has the minimum MSE.
